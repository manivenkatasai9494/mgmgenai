Show All Columns

```python
import pandas as pd

pd.set_option(
    "display.max_columns",
    None
)

df = pd.read_csv("hr_data.csv")

print(df.head())
```

Show All Rows

```python
import pandas as pd

pd.set_option(
    "display.max_rows",
    None
)

df = pd.read_csv("hr_data.csv")

print(df)
```

Print CSV Line by Line

```python
import pandas as pd

df = pd.read_csv("hr_data.csv")

for index, row in df.iterrows():

    print(row)

    print("=" * 50)
```

Print Row in One Line

```python
import pandas as pd

df = pd.read_csv("hr_data.csv")

for index, row in df.iterrows():

    print(row.to_dict())
```

Filter Salary Greater Than

```python
import pandas as pd

df = pd.read_csv("hr_data.csv")

filtered = df[
    df["salary"] > 100000
]

print(filtered)
```
